```python
# Define a function to compute binned color features  
def bin_spatial(img, size=(64,64)):
    # Use cv2.resize().ravel() to create the feature vector
    features = cv2.resize(img, size).ravel() 
    # Return the feature vector
    return features
```

```python
def color_hist(img, nbins=64, bins_range=(0, 256)):
    # Compute the histogram of the color channels separately
    channel1_hist = np.histogram(img[:,:,0], bins=nbins, range=bins_range)
    channel2_hist = np.histogram(img[:,:,1], bins=nbins, range=bins_range)
    channel3_hist = np.histogram(img[:,:,2], bins=nbins, range=bins_range)
    # Concatenate the histograms into a single feature vector
    hist_features = np.concatenate((channel1_hist[0], channel2_hist[0], channel3_hist[0]))
    # Return the individual histograms, bin_centers and feature vector
    return hist_features
```

```python
def colour_stuff(img):
    feature_image = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
    spacial=bin_spatial(feature_image)
    color=color_hist(feature_image)
    return np.concatenate((color,spacial))
ind = np.random.randint(0, len(car_images))
# Read in the image
image = car_images[ind]

# Plot the examples
features=colour_stuff(image)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.set_title('Example Car Image', fontsize=30)
ax2.plot(features)
ax2.set_title('YCrCb Binned', fontsize=30)
plt.savefig('output_images/bin.png', bbox_inches='tight')

ind = np.random.randint(0, len(notcar_images))
# Read in the image
image = notcar_images[ind]

# Plot the examples
features=colour_stuff(image)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.set_title('Example Non-car Image', fontsize=30)
ax2.plot(features)
ax2.set_title('YCrCb Binned', fontsize=30)
plt.savefig('output_images/noncar_bin.png', bbox_inches='tight')
    
```

```python
from skimage.feature import hog
def hog_image(img, orient, pix_per_cell, cell_per_block, feature_vec=True):
    features, hog_image = hog(img, orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=True, feature_vector=feature_vec)
    return features, hog_image

def hog_image_plain(img, orient, pix_per_cell, cell_per_block, feature_vec=True):
    features = hog(img, orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell), cells_per_block=(cell_per_block, cell_per_block), visualise=False, feature_vector=feature_vec)
    return features
ind = np.random.randint(0, len(car_images))
# Read in the image
image = car_images[ind]
gray = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
# Define HOG parameters
orient = 9
pix_per_cell = 4
cell_per_block = 4
# Call our function with vis=True to see an image output
features, hog_img = hog_image(gray[:,:,0], orient, 
                        pix_per_cell, cell_per_block,  feature_vec=False)


# Plot the examples
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.set_title('Example Car Image', fontsize=30)
ax2.imshow(hog_img, cmap='gray')
ax2.set_title('HOG Visualization', fontsize=30)
plt.savefig('output_images/hog.png', bbox_inches='tight')

ind = np.random.randint(0, len(notcar_images))
image = notcar_images[ind]
gray = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
features, hog_img = hog_image(gray[:,:,0], orient, pix_per_cell, cell_per_block,  feature_vec=False)


# Plot the examples
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.set_title('Example Non-Car Image', fontsize=30)
ax2.imshow(hog_img, cmap='gray')
ax2.set_title('HOG Visualization', fontsize=30)
plt.savefig('output_images/nonCar_hog.png', bbox_inches='tight')
```

Now putting it all together

```python
def extract_features(imgs, orient=9, pix_per_cell=4, cell_per_block=4, hog_channel=0):
    features=[]
    for img in imgs:
        feature_image = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
        if hog_channel == 'ALL':
            hog_features = []
            for channel in range(feature_image.shape[2]):
                hog_features.append(hog_image_plain(feature_image[:,:,channel], 
                                    orient, pix_per_cell, cell_per_block,  feature_vec=True))
            hog_features = np.ravel(hog_features)        
            print(hog_image_plain(feature_image[:,:,channel], orient, pix_per_cell, cell_per_block,  feature_vec=False).shape)
        else:
            hog_features = hog_image(feature_image_plain[:,:,hog_channel], orient, 
                        pix_per_cell, cell_per_block, feature_vec=True)
        spatial_features = bin_spatial(feature_image)
        # Apply color_hist() also with a color space option now
        hist_features = color_hist(feature_image)
        features.append(np.concatenate((spatial_features,hist_features,hog_features)))
    return features
```

```python
def find_car_candidates(img,ystart,ystop,scale,classifier,scaler,pix_per_cell=4,cell_per_block=4,orient=9,spatial_size=(64,64),hist_bins=64):
    #
    img_tosearch = img[ystart:ystop,:,:]
    ctrans_tosearch=cv2.cvtColor(img_tosearch, cv2.COLOR_BGR2YCrCb)
    if scale != 1:
        imshape = ctrans_tosearch.shape
        ctrans_tosearch = cv2.resize(ctrans_tosearch, (np.int(imshape[1]/scale), np.int(imshape[0]/scale)))
        
    ch1 = ctrans_tosearch[:,:,0]
    ch2 = ctrans_tosearch[:,:,1]
    ch3 = ctrans_tosearch[:,:,2]

    # Define blocks and steps as above
    nxblocks = (ch1.shape[1] // pix_per_cell)-1
    nyblocks = (ch1.shape[0] // pix_per_cell)-1 
    # 64 was the orginal sampling rate, with 8 cells and 8 pix per cell
    window = 64
    cells_per_step = 2  # Instead of overlap, define how many cells to step
    nblocks_per_window = window // pix_per_cell - cell_per_block + 1
    nxsteps = (nxblocks - nblocks_per_window) // cells_per_step
    nysteps = (nyblocks - nblocks_per_window) // cells_per_step
    
    # Compute individual channel HOG features for the entire image
    hog1 = hog_image_plain(ch1, orient, pix_per_cell, cell_per_block, feature_vec=False)
    hog2 = hog_image_plain(ch2, orient, pix_per_cell, cell_per_block, feature_vec=False)
    hog3 = hog_image_plain(ch3, orient, pix_per_cell, cell_per_block, feature_vec=False)
    search_rects=[]
    rects=[]
    for xb in range(nxsteps):
        for yb in range(nysteps):
            ypos = yb*cells_per_step
            xpos = xb*cells_per_step
            # Extract HOG for this patch
            hog_feat1 = hog1[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_feat2 = hog2[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_feat3 = hog3[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_features = np.concatenate([hog_feat1, hog_feat2, hog_feat3])
            xleft = xpos*pix_per_cell
            ytop = ypos*pix_per_cell
            # Extract the image patch
            subimg = cv2.resize(ctrans_tosearch[ytop:ytop+window, xleft:xleft+window], (64,64))
            spatial_features = bin_spatial(subimg, size=spatial_size)
            hist_features = color_hist(subimg, nbins=hist_bins)
            combined_features=[np.concatenate((spatial_features,hist_features,hog_features))]
            test_features = scaler.transform(combined_features) 
            test_prediction = classifier.predict(test_features)
            xbox_left = np.int(xleft*scale)
            ytop_draw = np.int(ytop*scale)
            win_draw = np.int(window*scale)
            p1 = (xbox_left, ytop_draw+ystart)
            p2 = (xbox_left+win_draw,ytop_draw+win_draw+ystart)
            rect = (p1, p2)
            search_rects.append(rect)
            if test_prediction[0] == 1:
                rects.append(rect)
    return rects, search_rects
```

```python
def find_car_candidates_at_scales(img, scales):
    # Parameters for find_cars
    ystart = 350
    ystop = 675
    spatial_size = (64,64)
    hist_bins = 64

    rects = []
    search_rects = []
    for scale in scales:
        results, search = find_car_candidates(img, ystart, ystop, scale, clf, X_scaler, 4, 4,9, spatial_size, hist_bins)
        rects.extend(results)
        search_rects.extend(search)
        
    return rects, search_rects

def draw_boxes(img, rects):
    draw_img = np.copy(img)
    for p1, p2 in rects:
        cv2.rectangle(draw_img, p1, p2, (255,0,0), 6)
    return draw_img

test_img=cv2.imread('test_images/test6.jpg')
rects, search_rects = find_car_candidates_at_scales(test_img, [1.5, 2, 2.5])
search_img = draw_boxes(test_img, search_rects)
out_img = draw_boxes(test_img, rects)

# Plot the examples
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(cv2.cvtColor(search_img, cv2.COLOR_BGR2RGB))
ax1.set_title('Search boxes', fontsize=30)
ax2.imshow(cv2.cvtColor(out_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Cars Found', fontsize=30)
plt.savefig('output_images/search.png', bbox_inches='tight')
```

```python
from scipy.ndimage.measurements import label

def filter_false_positives(boxes):
        min_tolerance = 75
        max_tolerance = 250
        box_list = []
        for bbox in boxes:
            if abs(bbox[1][0] - bbox[0][0]) >= min_tolerance and abs(bbox[1][1] - bbox[0][1]) <= max_tolerance:
                box_list.append(bbox)
        return box_list
    
def add_heat(heat_img,bboxs):
    for bbox in bboxs:
        heat_img[bbox[0][1]:bbox[1][1],bbox[0][0]:bbox[1][0]]+=1
    return heat_img

def apply_thresh(heat_img,thresh):
    heat_img[heat_img<=thresh]=0
    return heat_img

def get_labeled_bboxs(labels):
    bboxs=[]
    for car_number in range(1, labels[1]+1):
        nonzero = (labels[0] == car_number).nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        bboxs.append([bbox[0],bbox[1]])
    return bboxs

def identify_cars(img):
    rects, _ = find_car_candidates_at_scales(img, [1.5, 2, 2.5])
    heat_img=np.zeros_like(img[:,:,0]).astype(np.float)
    heat_img=add_heat(heat_img,rects)
    heat_img=apply_thresh(heat_img,2)
    heat_img = np.clip(heat_img, 0, 255)
    labels=label(heat_img)
    labels=get_labeled_bboxs(labels)
    labels=filter_false_positives(labels)
    result_img=draw_boxes(np.copy(img),labels)
    debug_img=draw_boxes(img,rects)
    return result_img,heat_img,debug_img

for image_file in glob.glob('test_images/*.jpg'):
    test_img = cv2.imread(image_file)
    result_img,heatmap_img,debug_img=identify_cars(test_img)
    f, (ax1, ax2,ax3) = plt.subplots(1, 3, figsize=(20,10))
    ax1.imshow(cv2.cvtColor(debug_img, cv2.COLOR_BGR2RGB))
    ax1.set_title('Boxes', fontsize=30)
    ax2.imshow(heatmap_img, cmap='hot')
    ax2.set_title('heatmap', fontsize=30)
    ax3.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
    ax3.set_title('Final', fontsize=30)
    plt.savefig('output_images/final_'+os.path.basename(image_file), bbox_inches='tight')
    
```
